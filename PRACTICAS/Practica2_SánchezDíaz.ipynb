**Paso 3. Detección de valores atípicos**

```
vacunaciones.hist(figsize=(14,14), xrot=45)
plt.show()
```

```
vacunaciones.describe()
```

```
variables = ['total_vaccinations',	'people_vaccinated','people_fully_vaccinated',	'daily_vaccinations_raw', 'daily_vaccinations',	'total_vaccinations_per_hundred', 'people_vaccinated_per_hundred', 'people_fully_vaccinated_per_hundred', 'daily_vaccinations_per_million']

for col in variables:
    sns.boxplot(col, data= vacunaciones)
    plt.show()
```

NOTAS: existen muy pocos outliers en los datos,sin embargo todas las variables tienen sesgo, en donde se pueden observar claramente los outliers son en las vacunaciones diarias.

```
vacunaciones.describe(include='object')
```

```
for col in vacunaciones.select_dtypes(include='object'):
    if vacunaciones[col].nunique()<27:sns.countplot(y=col, data=vacunaciones)
    plt.show()
```

```
for col in vacunaciones.select_dtypes(include=object):
    if vacunaciones[col].nunique() <27:
        display(vacunaciones.groupby(col).agg(['mean']))
```

```
plt.figure(figsize=(30,10))
plt.plot(vacunaciones[vacunaciones.country=='Mexico']['date'], vacunaciones[vacunaciones.country=='Mexico']['people_vaccinated'], color='green', marker='o')
plt.show()
```

**Paso 4. Identificación de relaciones entre variables**

```
vacunaciones.corr()
```

```
plt.figure(figsize=(14,14))
sns.heatmap(vacunaciones.corr(), cmap='RdBu_r', annot=True)
plt.show()
```
